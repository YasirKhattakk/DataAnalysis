```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

sales = pd.read_csv('sample_data/sales_data.csv', parse_dates=['Date'])
sales.head()
```

```python
sales['Customer_Gender'].value_counts().plot(kind='density', figsize=(14,6))
```

```python
sales['Customer_Gender'].value_counts().plot(kind='bar', figsize=(10,2))
plt.title('Customer Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python

gender_numeric = pd.get_dummies(sales['Customer_Gender'], prefix='Gender')


plt.figure(figsize=(14,6))
for column in gender_numeric.columns:
    gender_numeric[column].plot(kind='density', label=column)
plt.title('Gender Distribution (Numerical Conversion)')
plt.legend()
plt.show()
```
